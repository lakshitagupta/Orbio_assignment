```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
urls = [
    'https://www.amazon.com/dp/B0C7VBWHLQ',
    'https://www.amazon.com/dp/B0BCH68K6M',
    'https://www.amazon.com/dp/B0C9VWK5H5',
    'https://www.amazon.com/dp/B0C1J4MPBB',
]
```

```python
result_data = []
for url in urls : 
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        #title
        title_string = soup.find("span", attrs={"id": 'productTitle'}).get_text(strip=True)
        split_title = title_string.split("(")
        name = split_title[0].strip()
        model_number = split_title[1].split(')')[0].strip()
        
        #rating
        rating = soup.find("span", attrs={"class": 'a-icon-alt'}).get_text(strip=True)
        
        #customer_rating
        customer_rating = soup.find("span", attrs={"id":'acrCustomerReviewText'}).get_text(strip=True)
        
        #price
        price = soup.find("span", attrs={"class": 'a-offscreen'}).get_text(strip=True)
        
        #image
        image_text = (soup.find("div", class_="imgTagWrapper").find("img")['src'] 
                      if soup.find("div", class_="imgTagWrapper") 
                      else "Image link not found")
        
        #additional_data
        table = soup.find("table", class_="a-normal a-spacing-micro")
        rows = table.find_all("tr")
        li = []
        for i in rows:
            data=i.find_all("td")
            li.append([tr.text for tr in data])
            temp_dict = {}
            
            for ele in li:
                if ele[0] not in temp_dict:
                    temp_dict[ele[0].strip()] = ele[1].strip()
        scraped_data={
            "Name:" : name,
            "Model Number:" : model_number,
            "Rating:":rating,
            "Customer_Ratings":customer_rating,
            "Price":price,
            "Image Link: ":image_text,
            "Additional Information:":temp_dict
        }
        result_data.append(scraped_data)
        
    else:
        print("Failed to fetch data from:", urls)
```
